## Step 2 after creating the dataset:
This step focuses on analysis of the data

```python
import pandas as pd
import matplotlib.pyplot as plt

# Loading the data and showing basic info:
df = pd.read_csv('prices.csv')
df.head()
```

```python
# Conerting Date:
df['Date'] = pd.to_datetime(df['Date'])
df['Date']
```

```python
avg_prices = df.groupby('Item')['Price'].mean()
print('\nAverage Prices:\n', avg_prices)
```

```python
# Plot trends:
for item in df['Item'].unique():
    item_data = df[df['Item']== item]
    plt.plot(item_data['Date'], item_data['Price'], label=item)
plt.legend()
plt.title("Price Trends Over Time")
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()
```

## Step 3: Adding one "smart insight":

```python
# Monthly Change:
df["Price_Change"] = df.groupby('Item')['Price'].diff()

print('\nPrice Changes:\n', df)
```
